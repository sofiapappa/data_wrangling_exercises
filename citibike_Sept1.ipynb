```python
import csv

source_file = open('202009-citibike-tripdata.csv', 'r')
output_file = open('2020_09_01-citibike.csv', 'w')
```

```python
citibike_reader = csv.DictReader(source_file)
print(citibike_reader.fieldnames)
```

```python
output_writer = csv.DictWriter(output_file, fieldnames=citibike_reader.fieldnames) #the header should be the same as in citibike_reader
output_writer.writeheader()
```

```python
for a_row in citibike_reader:
    start_time = a_row['starttime']
    timelist = start_time.split(" ") 
    the_date = timelist[0]
    if the_date == "2020-09-01":
        output_writer.writerow(a_row)
output_file.close()
```
